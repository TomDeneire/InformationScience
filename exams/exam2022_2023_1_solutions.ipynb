### Question 1-3

Describe a real-world example of a situation where you should use Unicode normalization. 

### Answer 1-3

The Unicode [FAQ](https://www.unicode.org/faq/normalization.html) section describes normalization as follows:

> Unicode encodes some text elements both as precomposed forms and as combining character sequences. The precomposed form has a canonical decomposition that makes the two representations canonically equivalent. Normalizing a string essentially means consistently picking one of these equivalent encodings, that is, either all composed or all decomposed. By contrast, unnormalized data may contain both forms.

In other words, even though Unicode supplies unique code points for characters, conversely some characters can be represented by multiple encodings. For instance:

```
print('\u00C5')
print('\u212B')
print('\u0041\u030A')
```

All of these print `Å`.

A real-world application of Unicode normalization would be text (pre)processing, for instance a word count tool. A word contained an `Å` character should be normalized to just one encoding sequence *before counting* so that the word count includes all instances of the same word instead of only those with a specific encoding.

````python
import json

from urllib.request import urlopen
from urllib.error import HTTPError, URLError
import sys


MANIFEST_IDS = {
    "7775db2cb90285e316498d2818862b8389a068ca",
    "ff9a10521469941f251878b2a606e5d746a1d858",
    "a348540a300c940bbccc6ae5d1d8d7d5fdabcf6f",
}

IIIF_PREFIX = "https://anet.be/iiif/"
TRANSFORMATION = "/full/max/0/gray.tif"


def harvest(url: str) -> bytes:
    try:
        with urlopen(url) as query:
            return query.read()
    except HTTPError as HTTPerr:
        sys.exit(HTTPerr.code)
    except URLError as URLerr:
        sys.exit(str(URLerr))


def extract_IIIF_info(manifest_ids: set) -> dict:
    result = {}
    for identifier in MANIFEST_IDS:
        url = IIIF_PREFIX + identifier + "/manifest"
        iiif_reponse = harvest(url)
        manifest = json.loads(iiif_reponse)
        result[identifier] = {}
        canvas = manifest["items"][0]["items"][0]["items"][0]["body"]["id"]
        result[identifier] = canvas + TRANSFORMATION
    return result

print(extract_IIIF_info(MANIFEST_IDS))
````
