### Question 2-1

Write a function that takes this list of word tuples, calculates the Jaro-Winkler similarity, and returns the list ranked by decreasing similarity (see answer section). 

### Answer 2-1 

(model answer: +/- 10 lines of code excluding the Jaro-Winkler algorithm)

```python
def jaro_distance(s1, s2):
    """
    Function to calculate the Jaro Similarity of two strings
    """

    # If the strings are equal
    if (s1 == s2):
        return 1.0

    # Length of two strings
    len1 = len(s1)
    len2 = len(s2)

    if (len1 == 0 or len2 == 0):
        return 0.0

    # Maximum distance upto which matching is allowed
    max_dist = (max(len(s1), len(s2)) // 2) - 1

    # Count of matches
    match = 0

    # Hash for matches
    hash_s1 = [0] * len(s1)
    hash_s2 = [0] * len(s2)

    # Traverse through the first string
    for i in range(len1):

        # Check if there is any matches
        for j in range(max(0, i - max_dist),
                       min(len2, i + max_dist + 1)):

            # If there is a match
            if (s1[i] == s2[j] and hash_s2[j] == 0):
                hash_s1[i] = 1
                hash_s2[j] = 1
                match += 1
                break

    # If there is no match
    if (match == 0):
        return 0.0

    # Number of transpositions
    t = 0

    point = 0

    # Count number of occurances where two characters match but
    # there is a third matched character in between the indices
    for i in range(len1):
        if (hash_s1[i]):

            # Find the next matched character in second string
            while (hash_s2[point] == 0):
                point += 1

            if (s1[i] != s2[point]):
                point += 1
                t += 1
            else:
                point += 1

        t /= 2

    # Return the Jaro Similarity
    return ((match / len1 + match / len2
             + (match - t) / match) / 3.0)


def jaro_winkler(s1, s2):
    """
    Jaro-Winkler Similarity
    """

    jaro_dist = jaro_distance(s1, s2)

    # If the jaro Similarity is above a threshold
    if (jaro_dist > 0.7):

        # Find the length of common prefix
        prefix = 0

        for i in range(min(len(s1), len(s2))):

            # If the characters match
            if (s1[i] == s2[i]):
                prefix += 1

            # Else break
            else:
                break

        # Maximum of 4 characters are allowed in prefix
        prefix = min(4, prefix)

        # Calculate jaro winkler Similarity
        jaro_dist += 0.1 * prefix * (1 - jaro_dist)

    return jaro_dist


def rank(words: list) -> list:
    """
    Longer version
    """
    result = {}

    for pair in words:
        result[pair] = jaro_winkler(pair[0], pair[1])

    result_sorted = sorted(
        result.items(), key=lambda pair: pair[1], reverse=True)

    return [item[0] for item in result_sorted]


def rank2(words: list) -> list:
    """
    Shorter version
    """

    return sorted(
        words,
        key=lambda pair: jaro_winkler(pair[0], pair[1]),
        reverse=True)


WORDS = [
    ('house', 'hose'),
    ('card', 'cord'),
    ('same', 'same'),
    ('today', 'yesterday'),
    ('pit', 'tip')
]

print(rank(WORDS))
print(rank2(WORDS))

EXPECTED_OUTPUT = [
    ('same', 'same'),
    ('house', 'hose'),
    ('card', 'cord'),
    ('pit', 'tip'),
    ('today', 'yesterday')
]

assert rank(WORDS) == rank2(WORDS) == EXPECTED_OUTPUT
```

#### Background information

IIIF (in full "International Image Interoperability Framework") is a collection of JSON-based web APIs by which providers (such as libraries, musea, collections) make images available for the outside world.

The IIIF Presentation API offers all metadata about a IIIF resource (a book, a painting, a piece of art, ...) as a "manifest" file (JSON), e.g.:

[https://anet.be/iiif/7775db2cb90285e316498d2818862b8389a068ca/manifest](https://anet.be/iiif/7775db2cb90285e316498d2818862b8389a068ca/manifest)


#### Question

Write a function that takes a manifest identifier, harvests its JSON manifest file online and converts the following metadata into an XML structure: ``id`` (identifier), ``items > id`` (identifier of each item), ``summary > en`` and ``summary > nl`` (English summary). See the expected output in the code.

### Answer 2-2

(model answer: +/- 40 lines of code)

```python
import json
import lxml.etree
from urllib.request import urlopen
from urllib.error import HTTPError, URLError
import sys

ANET_PREFIX = "https://anet.be/iiif/"

def harvest(manifest: str) -> bytes:
    url = ANET_PREFIX + manifest + "/manifest"
    try:
        with urlopen(url) as query:
            return query.read()
    except HTTPError as HTTPerr:
        sys.exit(HTTPerr.code)
    except URLError as URLerr:
        sys.exit(str(URLerr))

def convert(manifest: str) -> str:
    response = harvest(manifest)
    manifest_object = json.loads(response)
    root = lxml.etree.Element("data")
    for key, value in manifest_object.items():
        if key == "id":
            identifier = lxml.etree.SubElement(root, "id")
            identifier.text = value
        if key == "items":
            items_element = lxml.etree.SubElement(root, "items")
            items = value
            for item in items:
                item_element = lxml.etree.SubElement(items_element, "id")
                item_element.text = item["id"]
        if key == "summary":
            for lg in ["nl", "en"]:
                summary = lxml.etree.SubElement(
                    root, "summary", language=lg)
                summary.text = value[lg][0]

    return lxml.etree.tostring(root).decode()

MANIFEST_ID = "7775db2cb90285e316498d2818862b8389a068ca"
test = convert(MANIFEST_ID)
print(test)

EXPECTED_OUTPUT = '''<data><id>https://anet.be/iiif/7775db2cb90285e316498d2818862b8389a068ca/manifest</id><items><id>https://anet.be/iiif/7775db2cb90285e316498d2818862b8389a068ca/canvasbase/00000001</id><id>https://anet.be/iiif/7775db2cb90285e316498d2818862b8389a068ca/canvasbase/00000002</id><id>https://anet.be/iiif/7775db2cb90285e316498d2818862b8389a068ca/canvasbase/00000003</id></items><summary language="nl">Aflaeten ... verleent aen het Broederschap van de H. ... Ursula ...</summary><summary language="en">Aflaeten ... verleent aen het Broederschap van de H. ... Ursula ...</summary></data>'''

assert test == EXPECTED_OUTPUT
```
